## Text Splitting from documents - RecursiveCharcter text splitters


This text splitters is the recommended one for generic text. It is parameterized by a list of characters. It tries to split
on them the chunks are small enough. The default list is ["\n\n", "\n", " ", ""]. This has the effect of trying to keep all the paragraphs( and then the sentences , and the words) together as long as possible, as those would generically seem to be the stringest semantically related pieces of text.

- How the text is split: by the list of characters
- How the chunk size is measured: by the number of characters

```python
## reading a pdf
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader('Pandas_PDF.pdf')
docs = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
final_documents = splitter.split_documents(docs)
final_documents
```

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Header Text Splitter Test</title>
</head>
<body>
    <h1>Welcome to the Homepage</h1>
    <p>This is a sample paragraph under H1.</p>

    <h2>About Us</h2>
    <p>We are a team of developers.</p>

    <h3>Our Mission</h3>
    <p>To build useful tools.</p>

    <h4>Team Members</h4>
    <ul>
        <li>Alice</li>
        <li>Bob</li>
    </ul>

    <h5>Contact</h5>
    <p>Email: example@example.com</p>

    <h6>Disclaimer</h6>
    <p>All information is for testing purposes.</p>

    <div>
        <h2>Another Section</h2>
        <p>Content inside a div block.</p>
    </div>
</body>
</html>
"""


```

```python
header_to_split_on=[
    ("h1","Header 1"),
    ("h2", "Header 2"),
    ("h3","Header 3")
]

html_splitter = HTMLHeaderTextSplitter(header_to_split_on)
html_header_splits = html_splitter.split_text(html_string)
html_header_splits
```
