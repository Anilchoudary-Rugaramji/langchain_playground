## Sentence Transformers on Hugging Face

Hugging Face sentence-transformers is a python framework for state-of-the-art sentence, text, and image embeddings. One of the embeddings models is used in the hugging face embeddingd class. We also have added an alias sentence trnasformer Embeddings for the user are more familiar with directly using that package.

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text = "This is a test document for hugging face embeddings testing"
query_result = embeddings.embed_query(text)
query_result
```
