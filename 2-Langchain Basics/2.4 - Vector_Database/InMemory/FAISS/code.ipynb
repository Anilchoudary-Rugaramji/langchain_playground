```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
embeddings.embed_query("hello AI")
```

## Similarity Search Techniques

### Cosine - similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
documents = ["Who is the Prime Minister of INDIA?",
"what is the capital of USA?",
"Who is the President of USA",
]
```

```python
my_query="Narendra Modi is the Prime Minister of"
```

```python
query_embedding= embeddings.embed_query(my_query)
query_embedding
```

```python
document_embedding= embeddings.embed_documents(documents)
document_embedding
```

```python
cosine_similarity([query_embedding],document_embedding)
```

### Euclidean Distance

```python
from sklearn.metrics.pairwise import euclidean_distances
```

```python
euclidean_distances([query_embedding],document_embedding)
```

```python
import faiss
from langchain_community.vectorstores import FAISS
from langchain_community.docstore.in_memory import InMemoryDocstore

faiss.IndexFlatL2()
```

```python
index=faiss.IndexFlatL2(384)
```

```python
index
```

```python
vector_store=FAISS(
    index=index, 
    embedding_function=embeddings,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={}

)
```

```python
vector_store.add_texts(["Ai is Future", "AI is powerful", "Dogs are cute"])
```

```python
vector_store.index_to_docstore_id
```

```python
result = vector_store.similarity_search("Tell me about AI", k=2)
result
```

```python
result[0].metadata
```

```python
result[0].page_content
```
